```python
import re
f = open('D:\data\失物之书.txt') #一本英文小说
book = f.read()
pattern = r',|\.|/|;|—|\[|\]|<|>|\?|:|"|\{|\}|”|\~|!|@|#|\$|%|\^|&|\(|\)|=|\\_|\+|，|。|、|【|】|·|！| |…|\n|\r|（|）'
bookfc = re.split(pattern,book)
#print(ws)
bookfc1 = [i.lower() for i in bookfc if i!=''] #为了方便同意转化成小写
stopwd = ['am','is','are','were','was','a','an','to','in','from','by','after','about','that','also','but','however','i','you','he','she',
       'me','him','her','of','on','his','for','hers','mine','my','your','yours','it','the','and','they','be'] #自创停词表
booktc = [j for j in bookfc1 if j not in stopwd] 
print(len(booktc)/len(bookfc1))
print(bookfc1[1:100])
```

```python
dic = {} #通过字典记录词频
for i in booktc:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1
res = sorted(dic.items(),key=lambda item:item[1],reverse=True) #词频从高到低的结果
res[:100] #可以看出david是主要任务，其他的不明显
```
